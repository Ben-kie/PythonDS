題目:<br>
讀取STOCK_DAY_0050_202009.csv串聯STOCK_DAY_0050_202010.csv，並且找出open小於close的資料


```python
import pandas as pd
```

```python
#讀取STOCK_DAY_0050_202009.csv、STOCK_DAY_0050_202010.csv
stock_202009_data = pd.read_csv('D10_STOCK_DAY_0050_202009.csv')
stock_202010_data = pd.read_csv('D10_STOCK_DAY_0050_202010.csv')
print(stock_202009_data)
print('\n')
stock_202010_data
```

```python
#串聯
stock_20200910_data = pd.concat([stock_202009_data,stock_202010_data], axis = 0, join = 'inner')
print(stock_20200910_data)

stock_20200910_data_reset_index = stock_20200910_data.reset_index(drop=True)            #將此合併檔案'stock_20200910_data'重新編列索引
print(stock_20200910_data_reset_index)

stock_20200910_data_date_index = stock_20200910_data.set_index('date')                         #編列stock_20200910_data''檔案索引，用'date'欄位當作索引
print(stock_20200910_data_date_index)
stock_20200910_data_date_index.iloc[0:3,:2]                                                                          #利用索引篩選資料
```

[Pandas索引官方文件](https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html)  
[Pandas索引教學](https://www.yiibai.com/pandas/python_pandas_indexing_and_selecting_data.html)  

```python
#找出open小於close的資料
print(stock_20200910_data_date_index.loc[stock_20200910_data_date_index.open < stock_20200910_data_date_index.close,['open','close']])    #取出符合條件的資料
print('\n')

print(stock_20200910_data_date_index.loc[stock_20200910_data_date_index.open < stock_20200910_data_date_index.close,].index)                #取出符合條件的索引
print('\n')

print(stock_20200910_data_date_index.loc[stock_20200910_data_date_index.open < stock_20200910_data_date_index.close,].index.tolist())     #將index轉成列表形式
```
